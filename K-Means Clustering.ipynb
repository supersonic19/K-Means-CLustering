```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=500,centers=5,random_state=3)
```

```python
plt.figure(0) 
plt.grid(True)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

### Tip: It is an unsupervised learning algorithm

```python
k = 5

color = ["green","blue","red","yellow","orange"]

clusters = {}  # Empty Dictionary

for i in range(k):
    
    center = 10*(2*np.random.random((X.shape[1],))-1)
    points = []
    
    cluster =     {         # Dictionary which contains center, the points and the color of the cluster
        'center':center,
        'points':points,
        'color':color[i]
        
    }
    clusters[i] = cluster
    
print(clusters)
    
```

```python
def  distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def assignPointToClusters():
    for ix in range(X.shape[0]):   # E Step
        dist = []
        curr_x = X[ix]    
        
        
        for kx in range(k):
            d = distance(curr_x,clusters[kx]['center'])
            dist.append(d)
        
        current_cluster = np.argmin(dist)  
        clusters[current_cluster]['points'].append(curr_x)
    
def updateClusters():
    
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        
        if pts.shape[0]>0:
            new_u = pts.mean(axis=0)
            clusters[kx]['center'] = new_u
            clusters[kx]['points'] = []          # Clears the points as the center has now changed, thus the points
                                                 # will also hsve to be updated

def plotClusters():
    
    for kx in range(k):
        
        pts = np.array(clusters[kx]['points'])    # Prints the points of the cluster
        
        try:
            plt.scatter(pts[:,0],pts[:,1])
        except:
            pass
        # plotting the cluster center 
        
        uk = clusters[kx]['center']
        plt.scatter(uk[0],uk[1],color = "black",marker="*")
```

```python
assignPointToClusters()
plotClusters()
```

```python
updateClusters()
assignPointToClusters()
plotClusters()
```
